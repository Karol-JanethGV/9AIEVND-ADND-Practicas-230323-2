*12.3 Funciones que mandan llamar otras funciones y que regresan el resultado de un procedimiento* En este caso utilizaremos una funcion predefinidas 
en Python denominada *replace* que nos permite sustituir letras en una cadena de texto.

*Definicion:*

```python
def sustituir_vocales(texto):
    texto = texto.replace("A","4")
    texto = texto.replace("a","4")

    texto = texto.replace("E","3")
    texto = texto.replace("e","3")

    texto = texto.replace("I","1")
    texto = texto.replace("i","1")

    texto = texto.replace("O","0")
    texto = texto.replace("o","0")

    texto = texto.replace("U","∪")
    texto = texto.replace("u","∪")

    return texto
```

*Llamado:*

```python
print(sustituir_vocales(saludar_nombre("Karol Janeth")))
```

**13. Limpieza de letras ilegibles**

**14. Rellenar datos vacios por un valor generico asignado**

Durante el proceso de limpieza y Trnasformacion de los datos, existe la posibilidad de corregir datos que no se encuentran registrados, para 
este caso usaremos una funcion de **pandas** llamada *fillna*, (sirve para rellenar los datos no diponibles (not avalaible) en este caso seran sustituidos por la palabra *Others*, en el dataframe original por eso se activa **inplace=true**

```python
def value(val):
    val=str(val)
    return val.replace('?','a')
df['START']=df['START'].apply(value)
df
```

17. **Eliminamos la ultima fila de totales**
La ultima fila de totales no es necesaria para la graficacion dado que podria generar error en la representacion de datos por ello la eliminamos del dataframe usando el comando drop y el numero de la posicion de la fila a eliminar encerrada entre corchetes []

*¿Que es?* este codigo calcula los percentiles 95% (a) y 5% (b) de la columna MILES para identificar valores extremos y luego intenta filtrar el DataFrame dejando solo los registros dentro de ese rango 

*¿Porque?* Porque los percentiles 5% y 95% se usan para eliminar outliers (valores demasiado bajos o altos) que pueden distorsionar análisis o modelos, pero en tu código haces dos asignaciones separadas a df3, así que la segunda reemplaza la primera en vez de combinar ambos filtros.

 18.**Eliminar los datos de los cuartiles menores**


```python
a=df['MILES'].quantile(0.95)
b=df['MILES'].quantile(0.05)
df3=df[df['MILES']>b]
df3=df[df['MILES']<a]
df3
```

19. **Calculo de la Desviacion Estandar**
    Para entender que tan variables o dispersas son las distancias de los viajes

```python
df3['STD']= df.MILES.std()
df3
```

20. **Estimar el valor Z-score o Puntuacion Z**
    
Con la finalidad de estimar la dispersion de los datos, es decir que tan variados o diferentes son, utilizamos esta formula esdadistica para poder
continuar con el analisis EDA (Exploratory Data Analisys)


```python
df3['Z']= (df.MILES-df.MILES.mean())/df.MILES.std()
df3
```

23. **Grafico de clasifiacion de Viajes por** **CATEGORIA**

```python
fig,ax = mat.subplots(figsize=(8,5))

grafica = sea.countplot(
    data=df,
    x='CATEGORY',
    hue='CATEGORY',
    palette='Set2',
    legend=False,
    ax=ax
)

#Titulo
ax.set_title ('Categorias')

#Agregar etiquetas con totales
for container in ax.containers:
    ax.bar_label(container)
    
mat.show()
```

25. **Grafico de lugar de Abordaje (START)**

```python
fig,ax = mat.subplots(figsize=(15,50))

#Ordenar alfabeticamente
orden =sorted(df['START'].dropna().unique())

grafica = sea.countplot(
    data=df,
    y='START',
    order=orden,
    hue='START',
    dodge=False,
    palette='viridis',
    legend=False,
    ax=ax
)

#Titulo
ax.set_title ('Start Location')

#Agregar totales
for container in ax.containers:
    ax.bar_label(container)
    
mat.show()
```
